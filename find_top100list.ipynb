```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint

def get_market_cap_top100():
    top_100 = {}
    for page in range(1, 3):  # Fetch data from the first two pages to get the top 100 items
        url = f'https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={page}'
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
        }
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, "html.parser")

        stocks = soup.select('.type_2 tbody tr')

        for stock in stocks:
            cells = stock.find_all('td')
            if len(cells) >= 7:
                name = cells[1].text.strip()
                code = cells[1].a['href'].split('=')[-1]
                top_100[name] = {"code": code}
    
    return top_100

def get_listed_shares(code):
    url = f'https://finance.naver.com/item/main.naver?code={code}'
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    # Extract the listed shares
    listed_shares_info = soup.find(text='상장주식수').find_next('td')
    if listed_shares_info:
        listed_shares = listed_shares_info.text.strip().replace(',', '')
        return listed_shares
    return None

top_100 = get_market_cap_top100()

# Add specific data for certain stocks
top_100.update({
    '카카오': {'code': '035720', 'listed_shares': '443326185',  'split_date': '2021-04-15', 'split_ratio': 5},
    '한미반도체': {'code': '042700', 'listed_shares': '96993634', 'split_date': '2022-04-06', 'split_ratio': 2}
})

for name, info in top_100.items():
    code = info['code']
    if code not in ['035720', '042700']:
        listed_shares = get_listed_shares(code)
        top_100[name]['listed_shares'] = listed_shares

pprint(top_100, sort_dicts=False)
```
